### Identify Outliers

```python
# Function to identify outliers
def find_outliers(feature):
    Q1 = feature.quantile(0.25)
    Q3 = feature.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = feature[(feature < lower_bound) | (feature > upper_bound)]
    return outliers

# Dictionary to hold outliers for each feature
outliers_dict = {}

# Identify outliers for each numerical feature
for feature in numerical_features:
    outliers = find_outliers(data[feature].dropna())
    outliers_dict[feature] = outliers

# Display the outliers
for feature, outliers in outliers_dict.items():
    print(f"Outliers in {feature}:")
    print(outliers)
    print("\n")
```

## 3.0. Clustering Analysis

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt
import seaborn as sns


# Select numerical features relevant to economic freedom
selected_features = ['2019 Score', 'Property Rights', 'Judical Effectiveness', 'Government Integrity',
                     'Tax Burden', "Gov't Spending", 'Fiscal Health', 'Business Freedom',
                     'Labor Freedom', 'Monetary Freedom', 'Trade Freedom', 'Investment Freedom ',
                     'Financial Freedom']

# Handle missing values by imputing with the mean
data[selected_features] = data[selected_features].fillna(data[selected_features].mean())

# Standardize the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data[selected_features])

# Apply K-means clustering
kmeans = KMeans(n_clusters=3, random_state=0)  # You can adjust the number of clusters
kmeans_labels = kmeans.fit_predict(scaled_data)

# Add K-means cluster labels to the dataframe
data['KMeans_Cluster'] = kmeans_labels

# Visualize K-means clustering results
plt.figure(figsize=(15, 8))
sns.scatterplot(x=scaled_data[:, 0], y=scaled_data[:, 1], hue=kmeans_labels, palette='viridis')
plt.title('K-means Clustering Results')
plt.xlabel(selected_features[0])
plt.ylabel(selected_features[1])
plt.legend(title='Cluster')
plt.show()

# Apply Hierarchical clustering
linked = linkage(scaled_data, method='ward')

# Plot the dendrogram
plt.figure(figsize=(15, 8))
dendrogram(linked, orientation='top', labels=data['Country Name'].values, distance_sort='descending', show_leaf_counts=True)
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Country')
plt.ylabel('Distance')
plt.show()
```

## 4.0. Principal Componnet Analysis

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt


# Select numerical features relevant to economic freedom
selected_features = ['2019 Score', 'Property Rights', 'Judical Effectiveness', 'Government Integrity',
                     'Tax Burden', "Gov't Spending", 'Fiscal Health', 'Business Freedom',
                     'Labor Freedom', 'Monetary Freedom', 'Trade Freedom', 'Investment Freedom ',
                     'Financial Freedom']

# Handle missing values by imputing with the mean
#data[selected_features] = data[selected_features].fillna(data[selected_features].mean())

# Standardize the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data[selected_features])

# Apply PCA
pca = PCA(n_components=0.95)  # Retain enough components to explain at least 95% variance
pca.fit(scaled_data)

# Determine the number of components
num_components = pca.n_components_
print(f"Number of components that explain at least 95% of variance: {num_components}")

# Transform the data using the identified number of principal components
pca_data = pca.transform(scaled_data)

# Plot the cumulative explained variance to visualize
plt.figure(figsize=(10, 6))
plt.plot(np.cumsum(pca.explained_variance_ratio_), marker='o')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('Explained Variance by Principal Components')
plt.axhline(y=0.95, color='r', linestyle='--')
plt.axvline(x=num_components, color='r', linestyle='--')
plt.grid()
plt.show()
```

## 5.0. Perform Clustering and reduce Dimension

```python

# Determine the optimal number of clusters using the Elbow method
wcss = []  # Within-cluster sum of squares
silhouette_scores = []

for i in range(2, 11):  # Test for 2 to 10 clusters
    kmeans = KMeans(n_clusters=i, random_state=0)
    kmeans.fit(pca_data)
    wcss.append(kmeans.inertia_)
    silhouette_scores.append(silhouette_score(pca_data, kmeans.labels_))

# Plot the Elbow method results
plt.figure(figsize=(12, 6))
plt.plot(range(2, 11), wcss, marker='o', label='WCSS')
plt.title('Elbow Method for Optimal Number of Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.legend()
plt.show()

# Plot the Silhouette scores
plt.figure(figsize=(12, 6))
plt.plot(range(2, 11), silhouette_scores, marker='o', color='r', label='Silhouette Score')
plt.title('Silhouette Scores for Optimal Number of Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
plt.legend()
plt.show()

# Based on the Elbow and Silhouette method, choose the optimal number of clusters
optimal_clusters = 3  # You can adjust this based on the plots

# Apply K-means with the optimal number of clusters
kmeans = KMeans(n_clusters=optimal_clusters, random_state=0)
kmeans_labels = kmeans.fit_predict(pca_data)

# Add K-means cluster labels to the dataframe
data['Cluster'] = kmeans_labels

# Visualize the clustering results in 2D PCA space
plt.figure(figsize=(12, 8))
sns.scatterplot(x=pca_data[:, 0], y=pca_data[:, 1], hue=kmeans_labels, palette='viridis', s=100)
plt.title('K-means Clustering Results on Reduced Dimensions')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(title='Cluster')
plt.show()

# Describe the clusters
cluster_summary = data.groupby('Cluster')[selected_features].mean()
print("Cluster Summary:")
print(cluster_summary)
```
